```python
import os
import pandas as pd

# Ruta a la carpeta
data_folder = '/Users/mac/Desktop/Proyecto_siniestralidad_barcelona/ANTIGUO Datasets limpios'
a√±os = [2014, 2015]
dataframes = {}

print("Cargando archivos .xltx...\n")

for year in a√±os:
    file_name = f'Copia de {year}_accidents_gu_bcn.xltx'
    file_path = os.path.join(data_folder, file_name)

    try:
        df = pd.read_excel(file_path)
        dataframes[year] = df
        print(f"‚úÖ '{file_name}' cargado correctamente.")
    except FileNotFoundError:
        print(f"‚ùå ERROR: El archivo '{file_name}' no se encontr√≥ en '{data_folder}'.")
    except Exception as e:
        print(f"‚ùå ERROR al cargar '{file_name}': {e}")

# Mostrar resumen
print("\n--- üìä Resumen de DataFrames cargados ---")
for year, df in dataframes.items():
    print(f"A√±o {year}: {len(df)} filas, {len(df.columns)} columnas")
```

```python
from pyproj import Transformer

# Transformer de UTM zona 31N (Espa√±a, Barcelona) a WGS84
transformer = Transformer.from_crs("EPSG:25831", "EPSG:4326", always_xy=True)  # X = Este, Y = Norte

for year in [2014, 2015]:
    df = dataframes[year]

    # Renombrar columnas si hace falta
    df = df.rename(columns={
        'Coordenada UTM (X)': 'UTM_X',
        'Coordenada UTM (Y)': 'UTM_Y'
    })

    # Convertir coordenadas
    lons, lats = transformer.transform(df['UTM_X'].values, df['UTM_Y'].values)

    # A√±adir nuevas columnas al DataFrame
    df['Longitud'] = lons
    df['Latitud'] = lats

    dataframes[year] = df  # Guardar cambios en el diccionario

    print(f"‚úÖ Coordenadas convertidas para {year}")
```

```python
import pandas as pd
import os

data_folder = '/Users/mac/Desktop/Proyecto_siniestralidad_barcelona/Datasets corregidos'
dataframes = {}

# Incluir desde 2014 hasta 2024
for year in range(2014, 2025):
     # Usamos .xlsx solo para 2016, el resto son .xls
    extension = 'xlsx' if year == 2016 else 'xls'
    file_name = f"Copia de {year}_accidents_gu_bcn.{extension}"
    file_path = os.path.join(data_folder, file_name)

    try:
        df = pd.read_excel(file_path)
        dataframes[year] = df
        print(f"‚úÖ '{file_name}' cargado correctamente.")
    except FileNotFoundError:
        print(f"‚ùå ERROR: El archivo '{file_name}' no se encontr√≥ en '{data_folder}'.")
    except Exception as e:
        print(f"‚ùå ERROR al cargar '{file_name}': {e}")

print("\n--- üìä Resumen de DataFrames cargados ---")
for year, df in dataframes.items():
    print(f"A√±o {year}: {len(df)} filas, {len(df.columns)} columnas")
```

```python
# Comprobaci√≥n nombre de columnas para unificarlos

print("\n--- üîé Nombres de columnas por a√±o ---")
for year, df in dataframes.items():
    print(f"\nüìÖ A√±o {year} ({len(df.columns)} columnas):")
    print(list(df.columns))
```

```python
# limpieza de datos

def limpiar_columnas(df):
    df = df.copy()

    # Eliminar espacios en blanco alrededor de los nombres de columnas
    df.columns = df.columns.str.strip()

    # Renombrar columnas inconsistentes
    df = df.rename(columns={
        'Num_postal_caption': 'Num_postal',
        'Num_postal ': 'Num_postal',
        'NK_Any': 'Any'
    })

    # Eliminar columnas redundantes o inconsistentes
    columnas_a_eliminar = ['Dia_setmana', 'Descripcio_tipus_dia']
    for col in columnas_a_eliminar:
        if col in df.columns:
            df.drop(columns=col, inplace=True)

    return df
```

```python
# Se aplica esta limpieza de columnas eliminando las redundantes y unificando nombres
for year in dataframes:
    dataframes[year] = limpiar_columnas(dataframes[year])
```

```python
print("\n--- üß™ Columnas tras limpieza ---")
columnas_referencia = None
for year, df in dataframes.items():
    columnas_actuales = set(df.columns)
    print(f"A√±o {year}: {len(df.columns)} columnas")
    if columnas_referencia is None:
        columnas_referencia = columnas_actuales
    else:
        diferencia = columnas_actuales.symmetric_difference(columnas_referencia)
        if diferencia:
            print(f"  ‚ö†Ô∏è Diferencias respecto al primer a√±o: {diferencia}")
        else:
            print("  ‚úÖ Coinciden con el primer a√±o")
```
